# IRIS DATASET 

Singular Value Decomposition (SVD) tries to identify attributes that account for the most variance in the dataset. Unlike PCA, SVD is an unsupervised method and does not use class labels.

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.decomposition import TruncatedSVD
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
```

```python
# Load the Iris dataset
iris = datasets.load_iris()
X_iris = iris.data
y_iris = iris.target
```

```python
# Split the Iris dataset into training and test sets
X_train_iris, X_test_iris, y_train_iris, y_test_iris = train_test_split(X_iris, y_iris, test_size=0.3, random_state=42)
```

```python
# Standardize the features for Iris dataset
sc_iris = StandardScaler()
X_train_std_iris = sc_iris.fit_transform(X_train_iris)
X_test_std_iris = sc_iris.transform(X_test_iris)
```

```python
# Apply SVD for Iris dataset
svd_iris = TruncatedSVD(n_components=2)
X_train_svd_iris = svd_iris.fit_transform(X_train_std_iris)
X_test_svd_iris = svd_iris.transform(X_test_std_iris)
```

```python
# Supervised machine learning model (Logistic Regression) without SVD for Iris dataset
lr_iris = LogisticRegression()
lr_iris.fit(X_train_std_iris, y_train_iris)
y_pred_iris = lr_iris.predict(X_test_std_iris)
print("Classification Report without SVD - Iris:")
print(classification_report(y_test_iris, y_pred_iris))
```

```python
# Supervised machine learning model (Logistic Regression) with SVD for Iris dataset
lr_svd_iris = LogisticRegression()
lr_svd_iris.fit(X_train_svd_iris, y_train_iris)
y_pred_svd_iris = lr_svd_iris.predict(X_test_svd_iris)
print("\nClassification Report with SVD - Iris:")
print(classification_report(y_test_iris, y_pred_svd_iris))

print("\n\n")
```

# Making heat map after SVD 

```python
import pandas as pd 
import seaborn as sns 
# Apply SVD
svd = TruncatedSVD(n_components=2)
X_svd = svd.fit_transform(X_iris)

# Convert SVD components to a DataFrame
df_svd = pd.DataFrame(data=X_svd, columns=['SVD1', 'SVD2'])

# Calculate the correlation matrix
corr_svd = df_svd.corr()

# Plot the correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_svd, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap after SVD - Iris')
plt.show()
```

# MNIST DATASET 

```python
# Load the MNIST dataset
digits = datasets.load_digits()
X_digits = digits.data
y_digits = digits.target
```

```python
# Split the MNIST dataset into training and test sets
X_train_digits, X_test_digits, y_train_digits, y_test_digits = train_test_split(X_digits, y_digits, test_size=0.3, random_state=42)
```

```python
# Standardize the features for MNIST dataset
sc_digits = StandardScaler()
X_train_std_digits = sc_digits.fit_transform(X_train_digits)
X_test_std_digits = sc_digits.transform(X_test_digits)
```

```python
# Apply SVD for MNIST dataset
svd_digits = TruncatedSVD(n_components=2)
X_train_svd_digits = svd_digits.fit_transform(X_train_std_digits)
X_test_svd_digits = svd_digits.transform(X_test_std_digits)
```

```python
# Supervised machine learning model (Logistic Regression) without SVD for MNIST dataset
lr_digits = LogisticRegression()
lr_digits.fit(X_train_std_digits, y_train_digits)
y_pred_digits = lr_digits.predict(X_test_std_digits)
print("Classification Report without SVD - MNIST:")
print(classification_report(y_test_digits, y_pred_digits))
```

```python
# Supervised machine learning model (Logistic Regression) with SVD for MNIST dataset
lr_svd_digits = LogisticRegression()
lr_svd_digits.fit(X_train_svd_digits, y_train_digits)
y_pred_svd_digits = lr_svd_digits.predict(X_test_svd_digits)
print("\nClassification Report with SVD - MNIST:")
print(classification_report(y_test_digits, y_pred_svd_digits))
```

The Accuracy after applying SVD decreases this shows that there is no need to apply SVD in MNIST dataset
